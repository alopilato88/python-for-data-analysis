# Built-In Data Structures, Functions, and Files

## Data Structures and Sequences 

We will start by working with tuples, lists, and dictionaries.

### Tuple

A tuple is a fixed-length, immutable sequence of Python objects which, once assigned, cannot be changed. In Python, you create a tuple using `()`. I don't believe there is an `R` equivalent of a tuple.

```python
tup = ("a", "b", "c")
a, b = ("cat", "dog"), (1, 2)
d = a + b
e = a * 2
```

```python
a, b = 1, 2
print(a)
print(b)
b, a = a, b # Swapping values
print(a)
print(b)
```

### Lists

In contrast with tuples, lists are variable length and their contents can be modified in place. Lists are mutable. You can define them using square brackets `[]` or using the list type function.

```python
a_list = ["a", 1, True]
print(a_list)

gen = range(10)
gen_list = list(gen)
print(gen)
print(gen_list)
```

You can also add items to a Python list using the `append` function and remove them using the `pop` function or `remove` function. 

```python
a_list.append("appended item")
print(a_list)
a_list.pop(3)
print(a_list)
a_list.remove(1)
print(a_list)
```

You can also use `in` to determine if a value is contained in a list.

```python
"a" in a_list
```

You can also slice lists similar to how you slice vectors in `R`.

```python
x = [1, 2, 3, 4, 5]
x[1:4]
x[:4]
x[1:]
x[-5]
```

### Dictionaries

A dictionary stores a collection of key-value pairs, where key and value are Python objects. Each key is associated with a value so that a value can be conveniently retrieved, inserted, modified, or deleted given a particular key. One approach for creating a dictionary is to use curly braces `{}` and colons to separate keys and values.

```python
empty_dict = {}
d1 = {"a": ["some value", "other value"], "b": [1, 2, 3]}
print(empty_dict)
print(d1)
```

## Functions

Functions are the primary and most important method of code organization and reuse in Python. As a rule of thumb, if you anticipate needing to repeat the same or very similar code more than once, it may be worth writing a reusable function. Functions can also help make your code more readable by giving a name to a group of Python statements.

Functions are declared with the `def` keyword and are followed by a block of code and optional `return` keyword.

```python
def my_func(a, b):
    x = a + b
    return x

output = my_func(1, 5)
print(output)
```

Functions can have *positional* arguments and *keyword* arguments. Positional arguments have a specific position in the function's arguments and require values for the function to work. Keyword arguments have default values and the user is not required to specify them when they call the function. Positional arguments have to come before keyword arguments.

```python
def func_1(a, b, c = 1):
    return a + b - c

# The function below will cause an error becuase the keyword argument, a,
# is followed by positional arguments.
def func_2(a = 1, b, c):
    return a + b - c

print(func_1(5, 5))
print(func_2(5, 5))
```

Functions can also return multiple values.

```python
def f():
    a = 1
    b = 2
    c = 3
    return a, b, c

x, y, z = f()

print(x)
print(y)
print(z)
```

Python also has support for annonymous functions, which are also called *lamda* functions.

```python
def func_sq(x):
    return x**2

x = func_sq(2)
print(x)

anon_fun_sq = lambda x: x**2
y = anon_fun_sq(2)
print(y)
```
